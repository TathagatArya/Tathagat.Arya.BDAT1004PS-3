```python
import pandas as x
import numpy as y
import matplotlib.pyplot as a
import matplotlib.cm as b
%matplotlib inline

data = x.read_csv('arcade-revenue-vs-cs-doctorates.csv')
revenue = data['Total Arcade Revenue (billions)'].values
doctorates = data['Computer Science Doctorates Awarded (US)'].values
fig, xy = a.subplots()
colors = b.rainbow(y.linspace(0, 1, len(data['Year'])))

for i in range(len(data['Year'])):
    xy.scatter(revenue[i], doctorates[i],color=colors[i])

xy.set_xlabel('Total Revenue (billions)')
xy.set_ylabel('Computer Science Doctorates Awarded (US)')
```
